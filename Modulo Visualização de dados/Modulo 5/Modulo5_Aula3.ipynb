```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")

filiais = ['Filial Norte', 'Filial Sul', 'Filial Centro', 'Filial Leste', 'Filial Oeste']
desempenho_meta = [25, -12, 40, -5, 18]

df_desempenho = pd.DataFrame({
    'Filial': filiais,
    'Desempenho': desempenho_meta
    })

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

cores_aleatorias = ['red', 'orange', 'yellow', 'green', 'pink']

axes[0].barh(df_desempenho['Filial'], df_desempenho['Desempenho'], color=cores_aleatorias)

axes[0].set_title('1. Cores aleatorias')
axes[0].set_xlabel('Desempenho vs Meta')
axes[0].set_ylabel('Filiais')


cores_estrategicas = ['#2e7d32' if valor >= 0 else "#c62828" for valor in df_desempenho['Desempenho']]

barras = axes[1].barh(df_desempenho['Filial'], df_desempenho['Desempenho'], color=cores_estrategicas)

axes[1].axvline(0, color='black', linestyle='--')

axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].spines['left'].set_visible('black')


axes[1].set_title('2. Cores Estratégicas')
axes[1].set_xlabel('Desempenho vs Meta')
axes[1].set_ylabel('Filiais')

plt.tight_layout()
plt.show()
```
