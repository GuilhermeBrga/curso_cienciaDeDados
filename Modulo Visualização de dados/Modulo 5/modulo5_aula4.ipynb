```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")


empresas = ['Empresa Líder', 'Nossa Empresa', 'Concorrente B']
share = [45, 42, 13]

df_share = pd.DataFrame({'Empresa': empresas, 'Quota': share})


fig, axes = plt.subplots(1, 2, figsize=(16, 6))

axes[0].bar(df_share['Empresa'], df_share['Quota'], color=['#757575', '#d32f2f', '#757575'], width=0.5)

axes[0].set_ylim(40,46)

axes[0].set_title('1. Escala Desproporcional')

axes[0].set_ylabel('Porcentagem')

barras = axes[1].bar(df_share['Empresa'], df_share['Quota'], color=['#b0bec5','#1e88e5','#b0bec5'])

axes[1].set_ylim(0, 55)

axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)

for barra in barras:
    altura = barra.get_height()
    axes[1].annotate(f'{altura}%', xy=(barra.get_x() + barra.get_width() / 2, altura), xytext=(0, 5), textcoords='offset points', ha='center', va='bottom', fontsize=11, fontweight='bold', color='#b0bec5')

axes[1].set_title('2. Escala Proporcional')

axes[1].set_xlabel('Empresa')

axes[1].set_ylabel('Porcentagem')

plt.show()
```
