```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
vendas = [120, 125, 130, 175, 180, 182]

df_vendas = pd.DataFrame({
    'Mês': meses,
    'Unidades Vendidas': vendas
    })

fig, axes = plt.subplots(1, 2, figsize=(16, 6))


axes[0].plot(df_vendas['Mês'], df_vendas['Unidades Vendidas'], color='red', linewidth=2.5)
axes[0].set_title('1. Vendas por Mês')
axes[0].set_xlabel('M')
axes[0].set_ylabel('UV')


axes[1].plot(df_vendas['Mês'], df_vendas['Unidades Vendidas'], color='blue', linewidth=2.5)
axes[1].set_title('2. Campanha de marketing em Abril gerou crescimento de 35% de vendas', fontsize=13, fontweight='bold', color='black', pad=20)

axes[1].set_xlabel('Meses do primeiro semestre', fontsize=11, color='black')
axes[1].set_ylabel('Unidades vendidas (Em milhares)', fontsize=11, color='black')

axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)

axes[1].annotate('Inicio da Campanha',
                 xy=('Abr', 175),
                 xytext=('Fev', 175),
                 arrowprops=dict(facecolor='darkblue', shrink=0.08, width=1.5, headwidth=6),
                 fontsize=10, fontweight='bold', color='darkblue')

plt.tight_layout()
plt.show
```
