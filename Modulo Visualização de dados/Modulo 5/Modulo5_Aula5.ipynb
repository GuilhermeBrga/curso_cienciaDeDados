```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
canais = ['E-commerce', 'Lojas Físicas', 'Revendedores', 'App Mobile', 'Corporate']
faturamento = [450, 320, 280, 190, 110]

df_canais = pd.DataFrame({'Canal': canais, 'Faturamento': faturamento})


fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.set_style('whitegrid')

axes[0].bar(df_canais['Canal'], df_canais['Faturamento'], color='#90a4ae')

axes[0].set_title('1. Faturamento por Canal (Notebook)')
axes[0].set_xlabel('Canal')
axes[0].set_ylabel('Faturamento')

axes[1].grid(False)

cores_slide = ['#1e88e5', '#b0bec5', '#b0bec5', '#b0bec5', '#b0bec5']

barras = axes[1].bar(df_canais['Canal'], df_canais['Faturamento'], color=cores_slide, width=0.55)

axes[1].set_title('2. Faturamento por Canal (Slide)', fontsize=16, fontweight='bold',color='#1e88e5', pad=25)
axes[1].set_xlabel('Canal', fontsize=13, fontweight='bold', color='#e7474f', labelpad=12)
axes[1].set_ylabel('Faturamento', fontsize=13, fontweight='bold', color='#e7474f', labelpad=12)

axes[1].tick_params(axis='x', labelsize=12, labelcolor='#37474f')
axes[1].tick_params(axis='y', labelsize=12, labelcolor='#37474f')

axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].spines['left'].set_visible(False)
axes[1].spines['bottom'].set_color('#cfd8dc')
axes[1].spines['bottom'].set_linewidth(2)

plt.tight_layout()
plt.show()
```
