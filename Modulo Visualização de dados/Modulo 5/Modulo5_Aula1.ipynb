```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
produtos = ['Prod A', 'Prod B', 'Prod C', 'Prod D', 'Prod E']
vendas = [340, 510, 280, 620, 450]

df_vendas = pd.DataFrame({
        'Produto': produtos,
        'Vendas': vendas
    })


fig, axes = plt.subplots(1, 2, figsize=(16, 6))


cores_poluidas = ['#ff5722', '#4caf50', '#00bcd4', '#9c27b0', '#ffeb3b']
barras1 = axes[0].bar(df_vendas['Produto'], df_vendas['Vendas'], color=cores_poluidas, edgecolor='black', linewidth=2)
axes[0].grid(True, linestyle='-', linewidth=1.5, color='purple', alpha=0.5)
axes[0].set_title('1. Grafico poluido')
axes[0].set_xlabel('Produtos')
axes[0].set_ylabel('Vendas')

for spine in axes[0].spines.values():
    spine.set_visible(True)
    spine.set_linewidth(2)

cores_limpas = ['#b0bec5', '#b0bec5', '#b0bec5', '#1e88e5', '#b0bec5']

barras2 = axes[1].bar(df_vendas['Produto'], df_vendas['Vendas'], color=cores_limpas)

axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].spines['bottom'].set_visible(False)
axes[1].spines['left'].set_visible('cfd8dc')

axes[1].grid(False)
axes[1].get_yaxis().set_visible(False)

for barra in barras2:
    altura = barra.get_height()
    axes[1].annotate(f'{altura}',
                     xy=(barra.get_x() + barra.get_width() / 2, altura),
                     xytext=(0, 3),
                     textcoords='offset points',
                     ha='center', va='bottom', color='#37474f')

axes[1].set_title('2. Grafico limpo')
axes[1].set_xlabel('Produtos')


plt.tight_layout()
plt.show()
```
