```python
# ✅ Célula de setup — execute esta primeiro!
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
```

### Subplots e GridSpec

Para múltiplos gráficos na mesma Figure, temos duas opções:

| | `plt.subplots()` | `GridSpec` |
|---|---|---|
| Uso | Grades simples e regulares | Layouts assimétricos |
| Complexidade | Baixa | Média |
| Flexibilidade | Limitada | Total |

```python
# Layout assimétrico com GridSpec
fig = plt.figure(figsize=(12, 5), constrained_layout=True)
gs  = gridspec.GridSpec(2, 3, figure=fig)

# Painel grande: ocupa 2 linhas na coluna 0 e 1
ax_main = fig.add_subplot(gs[:, 0:2])
ax_main.plot(np.linspace(0, 10, 200), np.sin(np.linspace(0, 10, 200)), '#2E4057', lw=2)
ax_main.set_title('Painel Principal', fontweight='bold')
ax_main.spines[['top', 'right']].set_visible(False)

# Painel pequeno superior direito
ax_top = fig.add_subplot(gs[0, 2])
ax_top.scatter(np.random.rand(30), np.random.rand(30), color='#048A81', s=40, alpha=0.7)
ax_top.set_title('Scatter', fontweight='bold')
ax_top.spines[['top', 'right']].set_visible(False)

# Painel pequeno inferior direito
ax_bot = fig.add_subplot(gs[1, 2])
ax_bot.hist(np.random.randn(200), bins=15, color='#54C6EB', edgecolor='white')
ax_bot.set_title('Histograma', fontweight='bold')
ax_bot.spines[['top', 'right']].set_visible(False)

fig.suptitle('Layout assimétrico com GridSpec', fontsize=14, fontweight='bold')
plt.show()
```
