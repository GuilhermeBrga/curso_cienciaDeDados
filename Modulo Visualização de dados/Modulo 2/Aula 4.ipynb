### Figure vs Axes

**Figure** = a tela inteira  
**Axes** = o painel onde o gráfico aparece

Uma Figure pode ter vários Axes. Um Axes pertence a uma única Figure.

```python
"""
FIGURE E AXES

"""


# 🔴 Jeito IMPLÍCITO (pyplot) — funciona, mas pode causar problemas
plt.figure(figsize=(6, 3))
plt.plot([1, 2, 3], [4, 2, 5])
plt.title('Pyplot implícito')
plt.show()

# 🟢 Jeito EXPLÍCITO (Orientado a Objetos) — recomendado
fig, ax = plt.subplots(figsize=(6, 3))
ax.plot([1, 2, 3], [4, 2, 5])
ax.set_title('Interface OO — explícita e segura')
plt.show()
```
