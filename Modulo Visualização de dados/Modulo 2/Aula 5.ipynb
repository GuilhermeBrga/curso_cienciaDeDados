```python
# ✅ Célula de setup — execute esta primeiro!
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
```

### Subplots e GridSpec

Para múltiplos gráficos na mesma Figure, temos duas opções:

| | `plt.subplots()` | `GridSpec` |
|---|---|---|
| Uso | Grades simples e regulares | Layouts assimétricos |
| Complexidade | Baixa | Média |
| Flexibilidade | Limitada | Total |

```python
# Grade simples com subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 6), constrained_layout=True)

titulos = ['Gráfico A', 'Gráfico B', 'Gráfico C', 'Gráfico D']
cores   = ['#2E4057', '#048A81', '#54C6EB', '#E76F51']

for ax, titulo, cor in zip(axes.flat, titulos, cores):
    x = np.linspace(0, 10, 100)
    ax.plot(x, np.sin(x + np.random.rand()), color=cor, linewidth=2)
    ax.set_title(titulo, fontsize=12, fontweight='bold')
    ax.spines[['top', 'right']].set_visible(False)

fig.suptitle('Grade 2×2 com subplots()', fontsize=14, fontweight='bold')
plt.show()
```
