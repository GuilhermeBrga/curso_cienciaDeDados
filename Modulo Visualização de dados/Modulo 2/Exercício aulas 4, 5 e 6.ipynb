```python
# ✅ Célula de setup — execute esta primeiro!
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
```

### 🏋️ Exercício

**Desafio:** Monte um layout com GridSpec assim:
- Coluna esquerda: um gráfico de linha que ocupa as **2 linhas**
- Coluna direita, linha 0: um **scatter plot**
- Coluna direita, linha 1: um **histograma**

Use `np.random` para os dados. O objetivo é acertar a **estrutura de layout**, não os dados.

```python
# 👇 Escreva sua solução aqui
fig = plt.figure(figsize=(12, 5), constrained_layout=True)
gs  = gridspec.GridSpec(2, 2, figure=fig)

#Gráfico 1
ax_1 = fig.add_subplot(gs[:, 0])
ax_1.plot(np.random.rand(50).cumsum(), color='blue')
ax_1.set_title('Gráfico 1')

#Gráfico 2
ax_2 = fig.add_subplot(gs[0, 1])
ax_2.scatter(np.random.rand(20), np.random.rand(20), color='orange')

#Gráfico 3
ax_3 = fig.add_subplot(gs[1, 1])
ax_3.hist(np.random.randn(100), bins=15, color='green')

plt.show()
```
