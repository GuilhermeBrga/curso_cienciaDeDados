```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = [18, 5]
```

```python
np.random.seed(42)
x = np.random.normal(0, 1, 10000)
y = x + np.random.normal(0, 1, 10000)

fig, axes = plt.subplots(1, 3)

axes[0].scatter(x,y, color='blue')
axes[0].set_title('1. Scatter Plot Tradicional')
axes[0].set_xlabel('Variável X')
axes[0].set_ylabel('Variável Y')

axes[1].scatter(x,y, color='blue', alpha=0.15)
axes[1].set_title('2. Scatter Plot Alpha')
axes[1].set_xlabel('Variável X')
axes[1].set_ylabel('Variável Y')

hb = axes[2].hexbin(x, y, gridsize=30, cmap='Purples', mincnt=1)
axes[2].set_title('3. Scatter Plot Hexbin')
axes[2].set_xlabel('Variável X')
axes[2].set_ylabel('Variável Y')

cb = fig.colorbar(hb, ax=axes[2], orientation='vertical')
cb.set_label('Count')

plt.tight_layout()
plt.show()
```
