```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out']
faturamento = [1050, 1080, 1110, 1090, 1140, 1160, 1150, 1190, 1220, 1250]

df_vendas = pd.DataFrame({
    'Mês': meses,
    'Faturamento': faturamento
    })

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

axes[0].plot(df_vendas['Mês'], df_vendas['Faturamento'], color='gray', linewidth=3)
axes[0].set_ylim(1000, 1500)
axes[0].set_title('1. Sem Base Zero')
axes[0].set_xlabel('Mês')
axes[0].set_ylabel('Faturamento')

axes[1].plot(df_vendas['Mês'], df_vendas['Faturamento'], color='gray', linewidth=3)
axes[1].set_ylim(0, 1500)
axes[1].set_title('2. Com Base Zero')
axes[1].set_xlabel('Mês')
axes[1].set_ylabel('Faturamento')

plt.show()
```
