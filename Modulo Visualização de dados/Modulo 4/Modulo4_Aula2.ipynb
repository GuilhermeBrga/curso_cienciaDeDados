```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [16, 5]

np.random.seed(42)
horas_estudos = np.random.uniform(1,10,100)
notas = 3.0 + 0.6 * horas_estudos + np.random.normal(0, 1.2, 100)

df_estudos = pd.DataFrame({
    'Horas de Estudo': horas_estudos,
    'Notas': notas
    })

fig, axes = plt.subplots(1, 2)


sns.regplot(data=df_estudos, x='Horas de Estudo', y='Notas', ax=axes[0], color='green', scatter_kws={'alpha': 0.6})
axes[0].set_title('1. Regplot Classico')

sns.regplot(data=df_estudos, x='Horas de Estudo', y='Notas', ax=axes[1], color='green', ci=None, scatter_kws={'alpha': 0.6})
axes[1].set_title('2. Regplot Sem Intervalo de Confianca')


plt.tight_layout()
plt.show()
```

```python
df_estudos['Curso Preparatório'] = np.where(df_estudos['Horas de Estudo'] * 0.7 + np.random.normal(0,2,100) > 4, 'Sim', 'Não')

g = sns.lmplot(data=df_estudos, x='Horas de Estudo', y='Notas', hue='Curso Preparatório', palette='Dark2', height=5, aspect=2)

g.fig.suptitle('3. Lmplot')

plt.show
```
