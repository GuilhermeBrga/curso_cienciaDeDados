```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [14, 5]


data_venda_texto = ['2026-03-15', '2026-01-10', '2026-04-05', '2026-02-20',
                    '2026-05-12', '2026-07-22', '2026-06-18', '2026-08-30']
vendas = [120, 85, 150, 110, 190, 210, 175, 250]


df_vendas = pd.DataFrame({
    'Data Recibo': data_venda_texto,
    'Faturamento': vendas
})

print("--- DATAFRAME INICIAL (Problema: Data como texto e desordenada) ---")
print(df_vendas)
print(f"Tipo atual da coluna 'Data Recibo': {df_vendas['Data Recibo'].dtype}\n")


df_vendas['Data Real'] = pd.to_datetime(df_vendas['Data Recibo'])
df_vendas = df_vendas.sort_values(by='Data Real')

print("--- DATAFRAME TRATADO e ORDENADO ---")
print(df_vendas)
print(f"Tipo correto da nova coluna: {df_vendas['Data Real'].dtype}\n")

sns.lineplot(data=df_vendas, x='Data Real', y='Faturamento', color='red', linewidth=2.5, marker='o', markersize=8)

plt.title('Evolução de faturamento ao longo do tempo')
plt.xlabel('Linha do tempo (Meses de 2026)')
plt.ylabel('Total de vendas')

plt.tight_layout()
plt.show()
```
