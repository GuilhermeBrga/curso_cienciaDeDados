```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")

np.random.seed(10)
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']

regiao_norte = [120, 150, 140, 180, 210, 250]
regiao_sul   = [90, 110, 130, 120, 160, 180]
regiao_centro = [200, 190, 220, 210, 240, 230]


df_vendas = pd.DataFrame({
    'Mês': meses * 3,
    'Faturamento': regiao_norte + regiao_sul + regiao_centro,
    'Região': ['Norte']*6 + ['Sul']*6 + ['Centro']*6
})


g = sns.FacetGrid(df_vendas, col='Região', hue='Região', palette='dark', height=4.5, aspect=1.2)

g.map(plt.plot, 'Mês', 'Faturamento', linewidth=2.5)

g.map(plt.fill_between, 'Mês', 'Faturamento', alpha=0.2)

g.set(ylim=(0, 300))

g.set_axis_labels('Meses de 2026', 'Faturamento')

g.fig.suptitle('Faturamento por região')

plt.show()
```
